# **nbconvert**

The `nbconvert` tool enables converting Google colab or Jupyer `.ipynb` notebook document into another static file format such as: HTML, LaTeX, PDF, Markdown, etc

# Method 1: Save notebook directly to HTML, then Print the HTML file to PDF.


*   Easy to use, simple input/output
*   Formatting is not perfect
*   No need for any external libraries or dependencies



```python
!jupyter nbconvert --to html 'Claude_FMP_Commodities_Tracker_(Starter).ipynb'
```
